```python
a = int(input('Birinci sayiyi girin: '))
b= int(input('Ikinci sayiyi girin: '))
print('***  Hesap Makinesi  ***\n\
------------------------')
print('Sayilar   ..:', a, 've', b)
print('Toplami   ..:', a + b)
print('Farki     ..:', a - b)
print('Carpimi   ..:', a * b)
print('Bölümü    ..:', a / b)
print('Mod(', a, '%', b, ')  ..:', a % b, sep='')
print('1. sayinin karesi    ..:', a ** 2)
print('1. sayinin karekökü  ..:', a ** 0.5)
```
