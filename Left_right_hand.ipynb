```python
left = {'a', 'q', 'z', 's', 'w', 'x', 'd', 'e', 'c', 'f', 'r', 't', 'g', 'v', 'b'}
right = {'j', 'u', 'y', 'h', 'm', 'n', 'k', 'i', 'l', 'o', 'l', 'ö', 'p', 'ı', 'ü'}
gir = input('Lütfen bir kelime yaziniz..: ').lower().strip()
liste = set(gir)
a = left.union(right)
if liste.issubset(left):
    print('Sol el kullanildi...')
elif liste.issubset(right):
    print('Sag el kullanildi...')
elif liste.issubset(a):
    print('Iki el de kullanildi...')
else:
    print('Gecersiz giris...')
```
