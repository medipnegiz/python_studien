ITALYANCA - TÜRKCE SÖZLÜK

*   Türkce - Italyanca, Italyanca - Türkce olarak arama
*   Ic ice sözlük formatinda olacak
*   Girdi kullanicidan alinacak
*   Menü olacak (ne yapmak istiyorsunuz diye soracak)
      1.   Arama yap
          *   Türkce - Italyanca
          *   Italyanca - Türkce
      2.   Ekleme
      3.   Silme
      4.   Listeleme
      5.   Güncelleme
      6.   Cikis
    
*   Devam et yada kal kismi olacak





```python
import ast
with open('sozluk.txt', 'r', encoding='utf-8') as s:
    data = s.read()
sozluk = ast.literal_eval(data)
def git(yol):
    global line
    line = yol
line = 10
while True:
    if line == 10:
        print('Yapmak istediginiz islemin kodunu listeden secerek giriniz..:')         
        print('[1] Arama yapma', '[2] Ekleme', '[3] Silme', '[4] Listeleme', '[5] Güncelleme', '[6] Cikis', sep='\n')   
        sor = input().strip()
        git(20)
    elif line == 20:   
        if sor == '1':
            print('Lütfen sözlük türünü seciniz..:')
            print('[1] Türkce - Italyanca', '[2] Italyanca - Türkce', sep='\n')
            arama = input().strip()
            if arama == '1':
                key = (input('Türcke bir renk, hayvan yada meyve ismi girin..: ')).lower().strip()
                a = sozluk['italyanca']['renk'].get(key, 0)
                b = sozluk['italyanca']['hayvan'].get(key, 0)
                c = sozluk['italyanca']['meyve'].get(key, 0)
                if a != 0:
                    print(f'{key} : ', a)
                    while True:
                        devam = (input('Aramaya devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                        if devam == 'yes':
                            git(20)
                            break
                        elif devam == 'no':
                            git(100)
                            break
                        else:
                          print('Gecersiz giris...')
                elif b != 0:
                    print(f'{key} : ', b)
                    while True:
                        devam = (input('Aramaya devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                        if devam == 'yes':
                            git(20)
                            break
                        elif devam == 'no':
                            git(100)
                            break
                        else:
                          print('Gecersiz giris...')
                elif c != 0:
                    print(f'{key} : ', c)
                    while True:
                        devam = (input('Aramaya devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                        if devam == 'yes':
                            git(20)
                            break
                        elif devam == 'no':
                            git(100)
                            break
                        else:
                          print('Gecersiz giris...')
                else:
                      print('Sözlükte bulunamadi...')
                      git(10)
            elif arama == '2':
                val = (input('Italyanca bir renk, hayvan yada meyve ismi girin..: ')).lower().strip()
                if val in sozluk['italyanca']['renk'].values():
                    key2 = [k for k, v in sozluk['italyanca']['renk'].items() if v == val]
                    print(val, ':', *key2)
                    while True:
                        devam = (input('Aramaya devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                        if devam == 'yes':
                            git(20)
                            break
                        elif devam == 'no':
                            git(100)
                            break
                        else:
                          print('Gecersiz giris...')
                elif val in sozluk['italyanca']['hayvan'].values():
                    key2 = [k for k, v in sozluk['italyanca']['hayvan'].items() if v == val]
                    print(val, ':', *key2)
                    while True:
                        devam = (input('Aramaya devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                        if devam == 'yes':
                            git(20)
                            break
                        elif devam == 'no':
                            git(100)
                            break
                        else:
                          print('Gecersiz giris...')
                elif val in sozluk['italyanca']['meyve'].values():
                    key2 = [k for k, v in sozluk['italyanca']['meyve'].items() if v == val]
                    print(val, ':', *key2)
                    while True:
                        devam = (input('Aramaya devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                        if devam == 'yes':
                            git(20)
                            break
                        elif devam == 'no':
                            git(100)
                            break
                        else:
                          print('Gecersiz giris...')
                else:
                    print('Sözlükte bulunumadi...')
                    git(10)
        elif sor == '2':
            git(30)
        elif sor == '3':
            git(40)
        elif sor == '4':
            git(50)
        elif sor == '5':
            git(60)
        elif sor == '6':
            git(100)
        else:
            print('Yanlis giris...')
            git(10)
    elif line == 30:
        ne = (input('Hangisini eklemek istiyorsunuz: renk, hayvan, meyve ..: ')).lower().strip()
        if ne == 'renk':
            tur = (input('ekleme yapilacak Türkce rengi girin..: ')).lower().strip()
            if tur in sozluk['italyanca']['renk']:
                print('Sözlükte böyle bir kelime mevcut...')
                git(30)
            else:
                ita = (input(f'{tur} renginin Italyanca karsiligini girin..: ')).lower().strip()
                sozluk['italyanca']['renk'][tur] = ita
                with open('sozluk.txt', 'w', encoding='utf-8') as yaz:
                    yaz.write(str(sozluk))
                print(f'{tur} kelimesi sözlüge eklendi... \n')
                for i in sozluk['italyanca']['renk'].items():
                    print(i)
                while True:
                    devam = (input('Eklemeye devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                    if devam == 'yes':
                        git(30)
                        break
                    elif devam == 'no':
                        git(100)
                        break
                    else:
                        print('Gecersiz giris...')
        elif ne == 'hayvan':
            tur = (input('ekleme yapilacak Türkce hayvani girin..: ')).lower().strip()
            if tur in sozluk['italyanca']['hayvan']:
                print('Sözlükte böyle bir kelime mevcut...')
                git(30)
            else:
                ita = (input(f'{tur} hayvaninin Italyanca karsiligini girin..: ')).lower().strip()
                sozluk['italyanca']['hayvan'][tur] = ita
                with open('sozluk.txt', 'w', encoding='utf-8') as yaz:
                    yaz.write(str(sozluk))
                print(f'{tur} kelimesi sözlüge eklendi... \n')
                for i in sozluk['italyanca']['hayvan'].items():
                    print(i)
                while True:
                    devam = (input('Eklemeye devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                    if devam == 'yes':
                        git(30)
                        break
                    elif devam == 'no':
                        git(100)
                        break
                    else:
                        print('Gecersiz giris...')
        elif ne == 'meyve':
            tur = (input('ekleme yapilacak Türkce meyveyi girin..: ')).lower().strip()
            if tur in sozluk['italyanca']['meyve']:
                print('Sözlükte böyle bir kelime mevcut...')
                git(30)
            else:
                ita = (input(f'{tur} meyvesinin Italyanca karsiligini girin..: ')).lower().strip()
                sozluk['italyanca']['meyve'][tur] = ita
                with open('sozluk.txt', 'w', encoding='utf-8') as yaz:
                    yaz.write(str(sozluk))
                print(f'{tur} kelimesi sözlüge eklendi... \n')
                for i in sozluk['italyanca']['meyve'].items():
                    print(i)
                while True:
                    devam = (input('Eklemeye devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                    if devam == 'yes':
                        git(30)
                        break
                    elif devam == 'no':
                        git(100)
                        break
                    else:
                        print('Gecersiz giris...')
        else:
            print('Gecersiz giris...')
            git(30)
    elif line == 40: 
        ne = (input('Hangisinden silme yapmak istiyorsunuz: renk, hayvan, meyve ..: ')).lower().strip()
        if ne == 'renk':
            tur = (input('silinecek Türkce rengi girin..: ')).lower().strip()
            if tur in sozluk['italyanca']['renk']:
                sozluk['italyanca']['renk'].pop(tur)
                with open('sozluk.txt', 'w', encoding='utf-8') as yaz:
                    yaz.write(str(sozluk))
                print(f'{tur} kelimesi sözlükten silindi...')
                for i in sozluk['italyanca']['renk'].items():
                    print(i)
                while True:
                    devam = (input('Silmeye devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                    if devam == 'yes':
                        git(40)
                        break
                    elif devam == 'no':
                        git(100)
                        break
                    else:
                        print('Gecersiz giris...')
            else:
                print('Bu kelime sözlükte yok...')
        elif ne == 'hayvan':
            tur = (input('silinecek Türkce hayvani girin..: ')).lower().strip()
            if tur in sozluk['italyanca']['hayvan']:
                sozluk['italyanca']['hayvan'].pop(tur)
                with open('sozluk.txt', 'w', encoding='utf-8') as yaz:
                    yaz.write(str(sozluk))
                print(f'{tur} kelimesi sözlükten silindi...')
                for i in sozluk['italyanca']['hayvan'].items():
                    print(i)
                while True:
                    devam = (input('Silmeye devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                    if devam == 'yes':
                        git(40)
                        break
                    elif devam == 'no':
                        git(100)
                        break
                    else:
                        print('Gecersiz giris...')
            else:
                print('Bu kelime sözlükte yok...')
        elif ne == 'meyve':
            tur = (input('silinecek Türkce meyveyi girin..: ')).lower().strip()
            if tur in  sozluk['italyanca']['meyve']:
                sozluk['italyanca']['meyve'].pop(tur)
                with open('sozluk.txt', 'w', encoding='utf-8') as yaz:
                    yaz.write(str(sozluk))
                print(f'{tur} kelimesi sözlükten silindi...')
                for i in sozluk['italyanca']['meyve'].items():
                    print(i)
                while True:
                    devam = (input('Silmeye devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                    if devam == 'yes':
                        git(40)
                        break
                    elif devam == 'no':
                        git(100)
                        break
                    else:
                        print('Gecersiz giris...')
            else:
                print('Bu kelime sözlükte yok...')
        else:
            print('Gecersiz giris...')
            git(40)
    elif line == 50: 
        ne = (input('Hangisini listelemek istiyorsunuz: renk, hayvan, meyve ..: ')).lower().strip()
        if ne == 'renk':
            print('Renk Listesi: ')
            for i in sozluk['italyanca']['renk'].items():
                print(i)
            while True:
                devam = (input('Listelemeye devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                if devam == 'yes':
                    git(50)
                    break
                elif devam == 'no':
                    git(100)
                    break
                else:
                    print('Gecersiz giris...')
        elif ne == 'hayvan':
            print('Hayvan Listesi:')
            for i in sozluk['italyanca']['hayvan'].items():
                print(i)
            while True:
                devam = (input('Listelemeye devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                if devam == 'yes':
                    git(50)
                    break
                elif devam == 'no':
                    git(100)
                    break
                else:
                    print('Gecersiz giris...')
        elif ne == 'meyve':
            print('Meyve Listesi: ')
            for i in sozluk['italyanca']['meyve'].items():
                print(i)
            while True:
                devam = (input('Listelemeye devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                if devam == 'yes':
                    git(50)
                    break
                elif devam == 'no':
                    git(100)
                    break
                else:
                    print('Gecersiz giris...')
        else:
            print('Böyle bir liste bulunamadi...')
            git(50)
    elif line == 60: 
        ne = (input('Hangi listede güncelleme yapmak istiyorsunuz: renk, hayvan, meyve ..: ')).lower().strip()
        if ne == 'renk':
            gun1 = input('Güncellemek istediginiz Türkce rengi girin..: ').lower().strip()
            if gun1 in sozluk['italyanca']['renk']:
                gun2 = input(f'{gun1} kelimesinin Italyanca karsiligini girin..: ').lower().strip()
                sozluk['italyanca']['renk'][gun1] = gun2
                with open('sozluk.txt', 'w', encoding='utf-8') as yaz:
                    yaz.write(str(sozluk))
                print(f'{gun1} kelimesinin Italyanca karsiligi {gun2} olarak güncellendi')
                while True:
                    devam = (input('Güncellemeye devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                    if devam == 'yes':
                        git(60)
                        break
                    elif devam == 'no':
                        git(100)
                        break
                    else:
                        print('Gecersiz giris...')
            else:
                print('Kelime bulunamadi...')      
        elif ne == 'hayvan':
            gun1 = (input('Güncellemek istediginiz Türkce hayvan ismini girin..: ')).lower().strip()  
            if gun1 in sozluk['italyanca']['hayvan']:
                gun2 = input(f'{gun1} kelimesinin Italyanca karsiligini girin..: ').lower().strip()
                sozluk['italyanca']['hayvan'][gun1] = gun2
                with open('sozluk.txt', 'w', encoding='utf-8') as yaz:
                    yaz.write(str(sozluk))
                print(f'{gun1} kelimesinin Italyanca karsiligi {gun2} olarak güncellendi')
                while True:
                    devam = (input('Güncellemeye devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                    if devam == 'yes':
                        git(60)
                        break
                    elif devam == 'no':
                        git(100)
                        break
                    else:
                        print('Gecersiz giris...')
            else:
                print('Kelime bulunamadi...')  
        elif ne == 'meyve':
            gun1 = (input('Güncellemek istediginiz Türkce meyve ismini girin..: ')).lower().strip()
            if gun1 in sozluk['italyanca']['meyve']:
                gun2 = input(f'{gun1} kelimesinin Italyanca karsiligini girin..: ').lower().strip()
                sozluk['italyanca']['meyve'][gun1] = gun2
                with open('sozluk.txt', 'w', encoding='utf-8') as yaz:
                    yaz.write(str(sozluk))
                print(f'{gun1} kelimesinin Italyanca karsiligi {gun2} olarak güncellendi')
                while True:
                    devam = (input('Güncellemeye devam etmek istiyor musunuz (yes or no) :')).lower().strip()
                    if devam == 'yes':
                        git(60)
                        break
                    elif devam == 'no':
                        git(100)
                        break
                    else:
                        print('Gecersiz giris...')
            else:
                print('Kelime bulunamadi...')
        else:
            print('Böyle bir liste bulunamadi...')
            git(60)
    elif line == 100:
        print('IYI GUNLER')
        break
```
